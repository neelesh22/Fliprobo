## Q.2 Scrape the details team India’s international fixtures from bcci.tv

```python
import selenium
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup

driver=webdriver.Chrome()
driver.get('https://www.bcci.tv/')
```

```python
Title=[]
title_tag=driver.find_elements_by_xpath("//strong[@class='fixture__name fixture__name--with-margin']")
for i in title_tag:
    title=i.text
    Title.append(title)
    
Series=[]
series_tag=driver.find_elements_by_xpath("//span[@class='u-unskewed-text fixture__tournament-label u-truncated']")
for i in series_tag:
    series=i.text
    Series.append(series)
    
Place=[]
place_tag=driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/span")
for i in place_tag:
    place=i.text
    Place.append(place)
    
```

```python
Date=[]
date_tag=driver.find_elements_by_xpath("//span[@class='fixture__datetime tablet-only']/strong")
for i in date_tag:
    date=i.text
    Date.append(date)
```

```python
Time=[]
time_tag=driver.find_elements_by_xpath("//span[@class='fixture__datetime tablet-only']")
for i in time_tag:
    time=i.text
    time2=time.split(" ")
    Time.append(time2[3])
Time[:3]   
```

```python
df=pd.DataFrame({'Title':Title,
                'Series':Series,
                'Place':Place,
                'Date':Date,
                'Time(IST)':Time})

df
```

## Q.5 Scrape the details of trending repositories on Github.com

```python
import selenium
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import numpy as np
import time
```

```python
driver=webdriver.Chrome()


driver.get('https://github.com/')

try:
    driver.find_element_by_xpath("//nav[@class='mt-0 px-3 px-lg-0 mb-5 mb-lg-0']/ul/li[4]").click()
    driver.find_element_by_xpath("//nav[@class='mt-0 px-3 px-lg-0 mb-5 mb-lg-0']/ul/li[4]/details/div/ul[2]/li[3]/a").click()
except NoSuchElementException as e:
    WebDriverWait(driver, 10).until(ec.visibility_of_element_located((By.XPATH,"//nav[@class='mt-0 px-3 px-lg-0 mb-5 mb-lg-0']/ul/li[4]"))).click()
    WebDriverWait(driver, 2).until(ec.visibility_of_element_located((By.XPATH,"//nav[@class='mt-0 px-3 px-lg-0 mb-5 mb-lg-0']/ul/li[4]/details/div/ul[2]/li[3]/a"))).click()

titles=[]
des = []
counts = [] 
langs = []
ref = []
a = []

urls=driver.find_elements_by_xpath("//article[@class='Box-row']/h1/a")
for i in urls:
    ref.append(str(i.get_attribute('href')))
    
for url in ref:
    driver.get(url)
    try:
        title = driver.find_element_by_xpath("//main[@id='js-repo-pjax-container']/div/div/div/h1")
        titles.append(title.text.replace('\n',''))
    except NoSuchElementException as e:
        titles.append("-")

    try:
        de = driver.find_element_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div[1]/div/p")
        des.append(de.text)
    except NoSuchElementException as e:
        des.append("-")

    try:
        count = driver.find_element_by_css_selector("#repo-content-pjax-container > div > div.gutter-condensed.gutter-lg.flex-column.flex-md-row.d-flex > div.flex-shrink-0.col-12.col-md-3 > div > div:Nth-Last-Child(2)")
        counts.append(count.text.split()[1])
    
    except Exception as e:
        counts.append("-")
        
    try:
        lang = driver.find_element_by_css_selector("#repo-content-pjax-container > div > div.gutter-condensed.gutter-lg.flex-column.flex-md-row.d-flex > div.flex-shrink-0.col-12.col-md-3 > div > div:Last-Child > div > ul").text.split('\n')
        for index, element in enumerate(lang):
            if index % 2 == 0:
                a.append(element)
        langs.append(str(a))
        a.clear()
    except Exception as e:
        langs.append("-")

df = pd.DataFrame({'Repository title' : titles ,'Repository description' : des ,'Contributors count':counts ,'Languages used': langs})

driver.close()
df
```

## Q.6 Scrape the details of top 100 songs on billboard.com

```python
import selenium
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup

driver=webdriver.Chrome()
driver.get('https://www.billboard.com')
```

```python
hot_100=driver.find_element_by_xpath("//ul[@class='header__subnav__list display--flex flex--left-center flex-md--center-center']/li[3]")
hot_100.click()
```

```python

Name=[]
name_tag=driver.find_elements_by_xpath("//span[@class='chart-element__information__song text--truncate color--primary']")
for i in name_tag:
    name=i.text
    Name.append(name)

Singer=[]
singer_tag=driver.find_elements_by_xpath("//span[@class='chart-element__information__artist text--truncate color--secondary']")
for i in singer_tag:
    singer=i.text
    Singer.append(singer)
    
Rank_last_week=[]
Rank_last_tag=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--last']")
for i in Rank_last_tag:
    last_week=i.text
    Rank_last_week.append(last_week)
    
Rank_peak=[]
peak_tag=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--peak']")
for i in peak_tag:
    peak=i.text
    Rank_peak.append(peak)
    
Weeks_on_chart=[]
Weeks_chart=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--week']")
for i in Weeks_chart:
    week=i.text
    Weeks_on_chart.append(week)
    
df=pd.DataFrame({'Song Name':Name,
                'Artist Name':Singer,
                'Last Week Rank':Rank_last_week,
                'Peak Rank':Rank_peak,
                'Weeks On Board':Weeks_on_chart})

df
```

```python
driver.get('https://www.naukri.com/hr-recruiters-consultants')
```

```python
search_btn=driver.find_element_by_xpath("//input[@class='sugInp']")
search_btn.send_keys("Data Science")

driver.find_element_by_xpath("//button[@id='qsbFormBtn']").click()
```

```python
Name=[]
name_tag=driver.find_elements_by_xpath("//span[@class='fl ellipsis']")
for i in name_tag:
    name=i.text
    Name.append(name)
    
Designation=[]
desi_tag=driver.find_elements_by_xpath("//span[@class='ellipsis clr']")
for i in desi_tag:
    designation=i.text
    Designation.append(designation)
    
Company=[]
company_tag=driver.find_elements_by_xpath("//p[@class='highlightable']/a[2]")
for i in company_tag:
    company=i.text
    Company.append(company)
    
Skills=[]
skill_tag=driver.find_elements_by_xpath("//div[@class='hireSec highlightable']")
for i in skill_tag:
    skills=i.text
    Skills.append(skills)
    
Location=[]
location_tag=driver.find_elements_by_xpath("//p[@class='highlightable']")
for i in location_tag:
    location=i.text
    location_fin=location.split('\n')
    if len(location_fin) == 4:
        Location.append(location_fin[3])
    else:
        Location.append('---')
        
df=pd.DataFrame({'Name':Name,
                'Designation':Designation,
                'Company':Company,
                'Skills they hire for':Skills,
                'Location':Location})

df
```

## Q.9 Scrape the details most watched tv series of all time from imdb.com.

```python
import selenium
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup

driver=webdriver.Chrome()
driver.get('https://www.imdb.com/list/ls095964455/')

Name=[]
Year=[]
Genre=[]
Run_time=[]
Ratings=[]
Votes=[]

name_tag=driver.find_elements_by_xpath("//h3[@class='lister-item-header']/a")
year_tag=driver.find_elements_by_xpath("//span[@class='lister-item-year text-muted unbold']")
genre_tag=driver.find_elements_by_xpath("//span[@class='genre']")
runtime_tag=driver.find_elements_by_xpath("//span[@class='runtime']")
rating_tag=driver.find_elements_by_xpath("//div[@class='ipl-rating-star small']/span[2]")
votes_tag=driver.find_elements_by_xpath("//p[@class='text-muted text-small']/span[2]")

for i in name_tag:
    name=i.text
    Name.append(name)
    
for i in year_tag:
    year=i.text
    Year.append(year)
    
for i in genre_tag:
    genre=i.text
    Genre.append(genre)

for i in runtime_tag:
    runtime=i.text
    Run_time.append(runtime)

for i in rating_tag:
    ratings=i.text
    Ratings.append(ratings)
    
for i in votes_tag:
    votes=i.text
    if votes != '|':
        Votes.append(votes)
        
df=pd.DataFrame({'Name':Name,
                'Year Span':Year,
                'Genre':Genre,
                'Run Time':Run_time,
                'Ratings':Ratings,
                'Votes':Votes})
df
```
